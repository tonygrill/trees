```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('/home/antonio/Documentos/arboles/datasets/titanic.csv', sep = ',')
titanic.tail()
```

```python
columns_to_drop = ['Name', 'Fare']
titanic.drop(columns_to_drop, axis='columns', inplace = True)
```

```python
titanic.columns =  ['Sirvived', 'Pclass', 'Sex', 'Age', 'Sibsp', 'Parch']
```

```python
titanic.head()
```

```python
titanic.shape
```

```python
titanic.dtypes
```

```python
titanic = pd.get_dummies(titanic, columns = ['Sex'], drop_first = True)
```

```python
titanic.dtypes
```

```python
titanic.rename(columns={'Sex_male': 'Sex'}, inplace=True)
```

```python
titanic.dtypes
```

```python
titanic = titanic[['Sirvived', 'Pclass', 'Sex', 'Age', 'Sibsp', 'Parch']]
titanic.head()
```

```python
titanic.rename(columns={'Sirvived': 'Survived'}, inplace=True)
titanic.head()
```

```python
X = titanic.drop('Survived', axis=1)
y = titanic.Survived
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=00000)
```

```python
# Arbol de decisión
from sklearn.tree import DecisionTreeClassifier

#creación del modelo
tree = DecisionTreeClassifier(max_depth=2, random_state=0000)
```

```python
# entrenamiento del modelo
tree.fit(X_train, y_train)
```

```python
#EVALUACIÓN DEL MODELO
y_train_predit = tree.predict(X_train)
y_test_predit = tree.predict(X_test)
```

```python
# Cálculo de métricas
from sklearn.metrics import accuracy_score

# Cálculo del Accuracy en Train
train_accuracy = accuracy_score(y_train, y_train_predit)

# Cálculo del Accuracy en test
test_accuracy = accuracy_score(y_test, y_test_predit)

print('Accuracy en Train: ', train_accuracy)
print('Accuracy en Test: ', test_accuracy)
```

```python
# Verficamos el Feature Importances
import seaborn as sns
importances = tree.feature_importances_
columns = X.columns

importance_df = pd.DataFrame({
    'Feature': columns,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

# Graficar la importancia de las características
plt.figure(figsize=(10, 6))
sns.barplot(data=importance_df, x='Importance', y='Feature', hue='Feature', palette='bright', edgecolor='black', linewidth=2)
plt.title('Importancia de cada Feature')
plt.show()
```
